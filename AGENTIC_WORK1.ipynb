```python
import gradio as gr
from langchain_groq import ChatGroq
from google.colab import userdata
```

```python
llm = ChatGroq(
    api_key=userdata.get("GROQ_API_KEY"),
    model="llama-3.3-70b-versatile",
    temperature=0.3
)
```

```python
def generate_explanation(topic):
    prompt = f"Explain {topic} in simple terms for students."
    response = llm.invoke(prompt)
    return response.content
```

```python
def generate_quiz(topic):
    prompt = f"Generate 5 multiple choice quiz questions about {topic}."
    response = llm.invoke(prompt)
    return response.content
```

```python
def generate_summary(topic):
    prompt = f"Give a short bullet point summary of {topic}."
    response = llm.invoke(prompt)
    return response.content
```

```python
def generate_content(topic, content_type):

    if content_type == "Explanation":
        return generate_explanation(topic)

    elif content_type == "Quiz":
        return generate_quiz(topic)

    elif content_type == "Summary":
        return generate_summary(topic)
```

```python
interface = gr.Interface(
    fn=generate_content,
    inputs=[
        gr.Textbox(label="Enter Topic", placeholder="Example: Stack, Binary Tree, Machine Learning"),
        gr.Radio(["Explanation", "Quiz", "Summary"], label="Select Content Type")
    ],
    outputs=gr.Textbox(label="Generated Educational Content"),
    title="рџ“љ AI Educational Content Generator",
    description="Enter a topic and generate explanation, quiz, or summary using AI."
)
```

```python
interface.launch()
```
